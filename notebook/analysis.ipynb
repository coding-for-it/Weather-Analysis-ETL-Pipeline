Insights from weather.csv

```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the weather data
df = pd.read_csv("../data/weather.csv") 
df.head()
```

```python
#Basic Info and Missing Values
print("Shape of dataset:", df.shape)
df.info()
df.isnull().sum()
```

```python
#Summary statistics
df.describe()
```

```python
#Temperature distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["temperature"], bins=15, kde=True, color="skyblue")
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Humidity distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["humidity"], bins=15, kde=True, color="salmon")
plt.title("Humidity Distribution")
plt.xlabel("Humidity (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Avg Temp by City
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="city", y="temperature", estimator="mean", ci=None, palette="viridis")
plt.title("Average Temperature by City")
plt.xticks(rotation=45)
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
#Temperature vs. Humidity
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="temperature", y="humidity", hue="city", palette="tab10")
plt.title("Temperature vs. Humidity by City")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
#How many cities are considered hot?
hot_cities = df[df['temperature'] > 32]["city"].nunique()
print(f"Number of cities where temperature > 32°C: {hot_cities}")
```
